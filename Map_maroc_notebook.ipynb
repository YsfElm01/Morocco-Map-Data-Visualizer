```python
start_col = df.columns.get_loc("Remarques")
# Supprimer toutes les colonnes à partir de "Remarques"
df = df.iloc[:, :start_col]
```

```python
#Espace Aménagé a 7 valeur null c'est le plus grand : on traite les valeurs null et aussi l'uniforme des valeurs
#valeur different de df["Espace Aménagé"].unique() : '5000 m²', '4690 m²', nan, 950,'SUPERFICIE COMMUN AVEC FF  2384'

#traitement sur les valeurs du col "Espace Aménagé"
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
import re

def extract_surface(value):
    if pd.isna(value):
        return np.nan

    # Convertir en string
    text = str(value).lower().strip()

    # Chercher un nombre (entier ou décimal)
    match = re.search(r"(\d+[.,]?\d*)", text)
    if match:
        # Remplacer la virgule par un point pour convertir en float
        num = match.group(1).replace(",", ".")
        return float(num)
    else:
        # Si aucun nombre n’est trouvé, on met NaN
        return np.nan

# Appliquer la fonction à la colonne
df["Espace Aménagé"] = df["Espace Aménagé"].apply(extract_surface)

# Vérification
print(df["Espace Aménagé"].head(15))


# Conversion de la colonne en numérique (au cas où il y aurait du texte)
df["Espace Aménagé"] = pd.to_numeric(df["Espace Aménagé"], errors="coerce")

# Créer l’imputer (ici stratégie = moyenne)
imputer = SimpleImputer(strategy="mean")

# Appliquer sur la colonne
df[["Espace Aménagé"]] = imputer.fit_transform(df[["Espace Aménagé"]])

# Vérification
print("\nValeurs nulles :", df["Espace Aménagé"].isna().sum())
```

```python
#supprimer les autre valeurs null :
df= df.dropna()
```

```python
#traitement sur plusieurs col :
import pandas as pd
import numpy as np
import re

# Supprimer les espaces au début et à la fin
df["region"] = df["region"].str.strip()

# -------------------------------
# 1️⃣ Milieu : supprimer les espaces et garder seulement "Urbain" et "Rural"
# -------------------------------
df["milieu"] = df["milieu"].str.strip()


#supprimer espace : 'CSP ZMAMRA    URBAIN '
df["nom_etablissement"] = (
    df["nom_etablissement"]
    .str.strip()                       # enlève espaces au début/fin
    .str.replace(r"\s+", " ", regex=True)  # remplace plusieurs espaces par un seul
)

# -------------------------------
# 2️⃣ Propriété : tout en minuscules puis capitaliser si besoin
# -------------------------------
df["propriete"] = df["propriete"].str.lower().str.strip()

# On peut aussi uniformiser certaines valeurs manuellement
df["propriete"] = df["propriete"].replace({
    "mjcc": "mjcc",
    "mjccc": "mjcc",
    "commune": "communal",
    "commune ": "communal",
    "domaine de l'etat": "domaine de l'etat",
    "domanial": "domanial",
    "privé alomrane": "privé alomrane",
    "al omrane": "alomrane",
    "mjs": "mjs",
    "ste idmaje skan": "ste idmaje skan",
    "ministére": "ministere",
    "mjcc": "mjcc"
})

# -------------------------------
# 3️⃣ Superficie : extraire le nombre et convertir en float
# -------------------------------
def clean_superficie(val):
    if pd.isna(val):
        return np.nan
    val_str = str(val).lower()
    # Chercher un nombre
    match = re.search(r"(\d+[.,]?\d*)", val_str)
    if match:
        return float(match.group(1).replace(",", "."))
    else:
        return np.nan

df["superficie_m2"] = df["superficie_m2"].apply(clean_superficie)

# -------------------------------
# 4️⃣ Date de rénovation : remplacer "***" par NaN, "aucune renovation" aussi
# -------------------------------
df["date_renovation"] = df["date_renovation"].replace({"***": np.nan, "aucune renovation ": np.nan})
df["date_renovation"] = pd.to_numeric(df["date_renovation"], errors="coerce")

# -------------------------------
# 5️⃣ Effectif : remplacer "***" par NaN et convertir en int
# -------------------------------
df["effectif"] = df["effectif"].replace({"***": np.nan})
df["effectif"] = pd.to_numeric(df["effectif"], errors="coerce")

# -------------------------------
# 6️⃣ Etat établissement : remplacer 1 par "Opérationnel"
# -------------------------------
df["etat_etablissement"] = df["etat_etablissement"].replace({1: "Opérationnel"})

# -------------------------------
# 7️⃣ Entité gestionnaire : supprimer les points
# -------------------------------
df["entite_gestionnaire"] = df["entite_gestionnaire"].str.replace(".", "", regex=False).str.upper().str.strip()

# -------------------------------
# 8️⃣ Titre foncier : nettoyer TF COMMUN et les // ou espaces inutiles
# -------------------------------
df["titre_foncier"] = df["titre_foncier"].astype(str).str.strip()
df["titre_foncier"] = df["titre_foncier"].replace({
    "TF COMMUN avec f f  93/91784": "TF_COMMUN",
    "//": np.nan,
    "Commune ": "COMMUNE",
    "COMMUNE ": "COMMUNE"
})

# -------------------------------
# 9️⃣ Etat batiment : convertir les nombres en string
# -------------------------------
etat_dict = {0: "Non équipé", 1: "Bon", 2: "Moyen", 3: "Mauvais", 4: "Médiocre"}
df["etat_batiment"] = df["etat_batiment"].replace(etat_dict)
# mettre tout en minuscules ou capitaliser
df["etat_batiment"] = df["etat_batiment"].str.capitalize()

# -------------------------------
# 🔟 Etat équipements : même logique
# -------------------------------
df["etat_equipements"] = df["etat_equipements"].replace(etat_dict)
df["etat_equipements"] = df["etat_equipements"].str.capitalize()
```

```python
#Traitment pour la compatibilite entre dataset et Map
#simplification des nom :
# Dictionnaire de correspondance
mapping = {
    "Ain sebaa hay mohammadi": "Mohammedia",
    "Ain Chock Hay Hassani Nouaceur": "Nouaceur",
    "Sidi Bennour": "MA-SIB",  # si tu veux remplacer par ce code
    # Les autres restent inchangés
}

# Appliquer le remplacement
df["province"] = df["province"].replace(mapping)

# Vérifier le résultat
print(df["province"].unique())
```

```python
#l'ajout d' id pour facilite la jointure du donnee du dataset avec la partie region dans map
# Dictionnaire des codes
province_codes = {
    "SETTAT": "MA-SET",
    "El Jadida": "MA-JDI",
    "Mohammedia": "MA-MOH",
    "Berrechid": "MA-BRR",
    "Benslimane": "MA-BES",
    "Nouaceur": "MA-NOU",
    "Sidi Bennour": "MA-SIB"
}

# Créer la colonne province_id en liant à province
df["province_id"] = df["province"].map(province_codes)

# Vérifier le résultat
print(df[["province", "province_id"]].head(10))
```
