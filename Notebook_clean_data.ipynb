```python
df["region"] = (
    df["region"]
    .astype(str)
    .str.strip()
    .str.replace("\n", " ")
)
# 2. Mapping 
df["region"] = df["region"].replace(mapping_region)
df = df.dropna(subset=["region"]).reset_index(drop=True)
```

```python
# #Traitement sur la colonne province

def nettoyer_province(val):
    # Convertir en string
    val = str(val).strip()

    # Supprimer les espaces multiples
    val = " ".join(val.split())

    # Liste des valeurs invalides
    valeurs_invalides = ["", "nan", "None", "NON DEFINI"]
    if val.lower() in valeurs_invalides:
        return "NON"

    return val

df["province"] = df["province"].apply(nettoyer_province)
```

```python
#Traitement sur la colonne lieu :
mapping_milieu = {
    "Rural": "Rural",
    "rurale": "Rural",
    "rurale ": "Rural",
    "Rurale": "Rural",

    "Urbain": "Urbain",
    "Urbain ": "Urbain",
    "urbain": "Urbain",
    "urbaine": "Urbain",
    "urbaine ": "Urbain",
}
def nettoyer_milieu(val):
    # Convert to string + trim
    val = str(val).strip().lower()

    # Normaliser
    if val in ["rural", "rurale"]:
        return "Rural"
    if val in ["urbain", "urbaine"]:
        return "Urbain"

    # Si ce n’est ni urbain ni rural → NON
    return "NON"
df["milieu"] = df["milieu"].apply(nettoyer_milieu)
```

```python
#Traitement sur : type centre :
import re

def nettoyer_type_centre(val):
    # Convertir en string
    if pd.isna(val):
        return "NON"

    val = str(val).strip()

    # Mettre tout en majuscule pour normaliser
    val_u = val.upper()

    # Centre sportif → CSP
    if "CENTRE SPORTIF" in val_u:
        return "CSP"

    # Complexe socio-sportif → CSP C
    if "COMPLEXE SOCIO-SPORTIF" in val_u:
        return "CSP C"

    # Si contient Type → extraire la lettre après "Type"
    if "TYPE" in val_u:
        # Extrait A/B/C : cherche une lettre après TYPE
        lettre = re.findall(r"TYPE\s*([A-Z])", val_u)
        if lettre:
            return f"CSP {lettre[0]}"
        else:
            return "CSP"   # si rien n'est trouvé

    #CSPI
    if val_u.startswith("CSPI"):
        # Ex. CSPI A / CSPI B / CSPI
        lettre = re.findall(r"CSPI\s*([A-Z])", val_u)
        if lettre:
            return f"CSPI {lettre[0]}"
        return "CSPI"

    # CSP (cas normaux)
    if val_u.startswith("CSP"):
        lettre = re.findall(r"CSP\s*([A-Z])", val_u)
        if lettre:
            return f"CSP {lettre[0]}"
        return "CSP"

    #Sinon → NON
    return "NON"
df["type_centre"] = df["type_centre"].apply(nettoyer_type_centre)
```

```python
#traitement sur etat_centre
mapping = {
    # Opérationnel
    'opérationnel': 'Opérationnel',
    'opérationnel ': 'Opérationnel',
    'achevé': 'Opérationnel',
    'prêt à être inauguré': 'Opérationnel',

    # En aménagement / construction
    "en cours des travaux d'aménagement": 'En aménagement',
    "en cours d'aménagement ": 'En aménagement',
    'en cours de construction': 'En aménagement',
    'en cours de réception ': 'En aménagement',
    "dans l'attente du budget de construction ": 'En aménagement',

    # En opérationnalisation / préparation
    'en cours d\'opérationnalisation': 'Quasi-opérationnel',
    "en cours d'opérationnalisation - prêt avec besoins/conditions": 'Quasi-opérationnel',
    "en cours d'opérationnalisation ": 'Quasi-opérationnel',

    # Non opérationnel
    'non opérationnel': 'Non opérationnel',
    'en arrêt': 'Non opérationnel',

    # Valeurs inconnues / aberrantes
    'mjcc': 'Inconnu',
    '1': 'Inconnu',
    'nan': 'Inconnu'
}

# Nettoyage simple : minuscules et suppression des espaces
df['etat_centre'] = df['etat_centre'].astype(str).str.lower().str.strip()

# Appliquer le mapping
df['etat_centre'] = df['etat_centre'].map(mapping).fillna('Inconnu')

```
